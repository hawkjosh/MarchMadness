```python
import requests
import pandas as pd
import json

url = "https://sdataprod.ncaa.com/?meta=GetContests_web&extensions={%22persistedQuery%22:{%22version%22:1,%22sha256Hash%22:%22f9d668706317cc187ecbe89591256e257b14af95432e36d4a68160dc78a7256c%22}}&variables={%22sportCode%22:%22MBB%22,%22division%22:1,%22seasonYear%22:2023,%22contestDate%22:%2202/20/2024%22,%22week%22:null}"
response = requests.get(url)
data = json.loads(response.text)

# Now, navigate through the JSON structure to find the data you need
# This step depends on the structure of your JSON
# For example, if the data is in a list under 'results' key
# data_needed = data['results']

# Convert that data into a DataFrame
# df = pd.DataFrame(data_needed)
```

```python
contests = data["data"]["contests"]

for contest in contests:
    if contest["teams"][0]["isWinner"] == False:
        print(
            contest["teams"][1]["name6Char"],
            contest["teams"][1]["score"],
            " - ",
            contest["teams"][0]["name6Char"],
            contest["teams"][0]["score"],
        )
    else:
        print(
            contest["teams"][0]["name6Char"],
            contest["teams"][0]["score"],
            " - ",
            contest["teams"][1]["name6Char"],
            contest["teams"][1]["score"],
        )
```

```python
contests = data["data"]["contests"]
rows = []

for contest in contests:
    row = {}
    row["date"] = contest[
        "startDate"
    ]  # replace 'contestDate' with the actual key for the date in your data
    if contest["teams"][0]["isWinner"] == False:
        row["team1"] = contest["teams"][1]["name6Char"]
        row["team1 score"] = contest["teams"][1]["score"]
        row["team2"] = contest["teams"][0]["name6Char"]
        row["team2 score"] = contest["teams"][0]["score"]
    else:
        row["team1"] = contest["teams"][0]["name6Char"]
        row["team1 score"] = contest["teams"][0]["score"]
        row["team2"] = contest["teams"][1]["name6Char"]
        row["team2 score"] = contest["teams"][1]["score"]
    rows.append(row)

df = pd.DataFrame(rows)
```
