```python
historicYears = [
    "2014",
    "2015",
    "2016",
    "2017",
    "2018",
    "2019",
    "2020",
    "2021",
    "2022",
    "2023",
]


def runHistoric(dirName, fileEnd, stats, years=historicYears):
    for year in years:
        df = compileTables(statYear=year, statDay="05-01", stats=stats)
        if not os.path.exists(dirName):
            os.makedirs(dirName)
        df.to_csv(
            f"{dirName}/{str(int(year[:4]) - 1)}-{year[2:4]}_{fileEnd}.csv",
            index=False,
        )


def runCurrent(dirName, fileEnd, stats, year="2024", day="05-01"):
    df = compileTables(statYear=year, statDay=day, stats=stats)
    if not os.path.exists(dirName):
        os.makedirs(dirName)
    df.to_csv(
        f"{dirName}/{str(int(year[:4]) - 1)}-{year[2:4]}_{fileEnd}.csv",
        index=False,
    )
```
