```python
def renameFiles(dir, filename):
    files = os.listdir(dir)
    for file in files:
        name = os.path.splitext(file)[0]
        year = name[-2:]
        
        os.rename(f"{dir}/{file}", f"{dir}/{filename}{year}.csv")
```

```python
def trnyMerge(mergeData, mergeDir, dropCols=None, renameCols=None):
    for data in mergeData:
        leftDf = data["leftDf"]
        rightDf = data["rightDf"]
        saveName = data["saveName"]

        if dropCols:
            rightDf.drop(columns=dropCols, inplace=True)

        dfMerge = pd.merge(leftDf, rightDf, on="team")

        if renameCols:
            dfMerge.rename(columns=renameCols, inplace=True)

        dfMerge.to_csv(f"{mergeDir}/{saveName}", index=False)
```
