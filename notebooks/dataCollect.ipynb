```python
from data_tools.dataCollecting import compileTables
```

```python
years = [
    "2014",
    "2015",
    "2016",
    "2017",
    "2018",
    "2019",
    "2020",
    "2021",
    "2022",
    "2023",
    "2024",
]


def runAllDataCollection(hist_dir, years, names, abbrevs, file_end):
    for year in years:
        df = compileTables(statYear=year, statNames=names, statAbbrevs=abbrevs)
        if year == "2024":
            df.to_csv(
                f"../data/raw/CurrYearData/{str(int(year[:4]) - 1)}-{year[2:4]}_{file_end}.csv",
                index=False,
            )
        else:
            df.to_csv(
                f"../data/raw/HistoricData/{hist_dir}/{str(int(year[:4]) - 1)}-{year[2:4]}_{file_end}.csv",
                index=False,
            )


def runCurrentDataCollection(names, abbrevs, file_end):
    df = compileTables(statYear="2024", statNames=names, statAbbrevs=abbrevs)
    df.to_csv(
        f"../data/raw/CurrYearData/{str(int('2024'[:4]) - 1)}-{'2024'[2:4]}_{file_end}.csv",
        index=False,
    )
```
