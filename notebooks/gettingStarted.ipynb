## NCAA Division I Men's Basketball Teams

This dataset includes basic identification data for all NCAA Division I Men's Basketball programs.

* `schoolName` - string value indicating team's school/university name
* `teamName` - string value indicating team's name
* `trTeamName` - string value indicating team's name from collected teamrankings stats (using this as universal)
* `kpTeamName` - string value indicating team's name from collected kenpom stats
* `teamMascot` - string value indicating team's mascot
* `teamConf` - string value indicating team's conference name
* `teamConfShort` - string value indicating team's shortened conference name (using this as universal)
* `kpTeamConf` - string value indicating team's conference name from kenpom stats
* `teamRegion` - string value indicating team's geographical region
* `teamState` - string value indicating team's state
* `privateFlag` - integer value indicating if team's school is private (1) or not (0)
* `hbcuBlackFlag` - integer value indicating if team's school is HBCU (1) or not (0)
* `currDivIFlag` - integer value indicating if team is currently division I (1) or not (0)

To select and view a single column from a DataFrame, use bracket notation with the column name, returns a pandas Series

```python
team_mascots = teams["teamMascot"]
team_mascots.head(10)
```

To select and view multiple specific columns from a DataFrame, use bracket notation with a list of column names, returns a pandas DataFrame

```python
teams_and_mascots = teams[["teamName", "teamMascot"]]
teams_and_mascots.tail(10)
```

To filter specific rows from a DataFrame, pass a conditional inside the bracket notation

```python
private_school_teams = teams[teams["privateFlag"] == 1]
private_school_teams.head()
```

NOTE: Filtering by specific row can also be done by first selecting a column, then passing in the conditional based on that column to bracket notation

```python
private_col = teams["privateFlag"]
private_schools = teams[private_col == 1]
private_schools.head()
```

```python
# create a bar chart of number of teams by geographic region
teams["teamRegion"].value_counts().plot(
    kind="bar",
    figsize=(10, 5),
    title="Teams by Geographic Region",
    xlabel="Geographic Region",
    ylabel="Number of Teams",
)
```

```python
# create a bar chart of number of teams by state
teams["teamState"].value_counts().plot(
    kind="bar",
    figsize=(10, 5),
    title="Teams by State",
    xlabel="State",
    ylabel="Number of Teams",
)
```
