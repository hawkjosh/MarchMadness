```python
def runEndRegSeasonCollect(dirName, fileEnd, stats):
    df = endRegSeasonUpdate(stats)
    if not os.path.exists(dirName):
        os.makedirs(dirName)
    df.to_csv(
        f"{dirName}/2023-24_{fileEnd}.csv",
        index=False,
    )
```
